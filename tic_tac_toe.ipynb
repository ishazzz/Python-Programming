```python
import numpy as np

# Create the board
def create_board():
    return np.array([[' ', ' ', ' '],
                     [' ', ' ', ' '],
                     [' ', ' ', ' ']])

# Display the board
def disp():
    for row in board:
        print(" | ".join(row))
    print()

# Take user input
def u_input(player):
    while True:
        row = int(input(f"Player {player}, enter row (0, 1, or 2): "))
        col = int(input(f"Player {player}, enter column (0, 1, or 2): "))

        if row < 0 or row > 2 or col < 0 or col > 2 or board[row][col] != ' ':
            print("Invalid move! Try again.")
        else:
            return row, col

# Check game status
def game_status(player, row, col):
    board[row][col] = player

    # Check rows and columns
    for i in range(3):
        if all(board[i, :] == player) or all(board[:, i] == player):
            disp()
            print(f"Player {player} wins!")
            return True

    # Check diagonals
    if all(np.diag(board) == player) or all(np.diag(np.fliplr(board)) == player):
        disp()
        print(f"Player {player} wins!")
        return True

    return False

# ---------------- MAIN PROGRAM ----------------
board = create_board()
current_player = 'X'

for _ in range(9):
    print("\nCurrent Board:")
    disp()

    row, col = u_input(current_player)

    if game_status(current_player, row, col):
        break

    current_player = 'O' if current_player == 'X' else 'X'

print("\nGame Over.")
```
